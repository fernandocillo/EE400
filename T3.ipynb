```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
w = np.linspace(-5,5,10000)
C = 0.05 #estabelecendo C=0.05 F

for r in range(0,5):   #variando o valor de R
  R = 10**r            #aumentando a variação de R sem poluir o gráfico
  g = 1/(1+(1j*w*R*C)) #ganho   
  g= abs(g)
  plt.plot(w, g, label='R= '+str(R)+' Ω',linewidth=2)  #plot do gráfico

#detalhes do gráfico
plt.legend(loc='upper right',fontsize='large')
plt.xlabel('Frequência angular', fontsize=14)
plt.ylabel('Ganho',fontsize=14)
plt.savefig('ganhoR.png') 
plt.title('Ganho do circuito RC \n em função de w e variando R', fontsize=20)    
plt.show()
```

```python
w = np.linspace(-5,5,10000)
R = 500 #estabelecendo R=500 ohm

for c in range(0,5):   #variando o valor de C
  C = 1/10**(c)         #aumentanbdo a variação de C sem poluir o gráfico
  g = 1/(1+(1j*w*R*C)) #ganho
  g= abs(g)
  plt.plot(w, g, label='C= '+str(C)+' F',linewidth=2)  #plot do gráfico

#detalhes do gráfico
plt.legend(loc='upper right',fontsize='large')
plt.xlabel('Frequência angular', fontsize=14)
plt.ylabel('Ganho',fontsize=14)
plt.savefig('ganhoC.png') 
plt.title('Ganho do circuito RC \n em função de w e variando C', fontsize=20)    
plt.show()
```

```python
#inicializando
t = np.linspace(-10,10,10000)
x=np.zeros_like(t)

plt.figure(figsize=(10,5)) #tamanho da figura plotada

#Série de Fourier com 1000 termos
for m in range(1,1000):
  s = 1/(-1j*m*np.pi)*(np.e**(-1j*m*np.pi)-1)*np.e**(1j*m*t)
  x = x + s

#detalhes do gráfico
plt.plot(t, x, label='Vin(t)')
plt.savefig('Vin.png')
plt.xlabel('Tempo (s)', fontsize=14)
plt.ylabel('Tensão de entrada',fontsize=14)
plt.title('Onda quadrada com período 2π', fontsize=20) 
plt.legend()
plt.show()
```

```python
#iniciando as variáveis que utilizaremos
t = np.linspace(-10,10,10000)
w = np.linspace(-10,10,10000)
C=0.3
R=10
out=0

#criando os vetores que serão utilizados para armazenar
#os valores da Série de Fourier
x=np.zeros_like(t)
saida=np.zeros_like(t)

plt.figure(figsize=(10,5)) #tamanho da figura plotada

#Série de Fourier com 100 termos
for m in range(1,100):
  s = 1/(-1j*m*np.pi)*(np.e**(-1j*m*np.pi)-1)*np.e**(1j*m*t) #onda quadrada
  x = x + s #acúmulo do valor de todos os termos
  ganho = 1/(1+1j*m*R*C) #expressão do ganho
  g =ganho*s #aplicando o ganho termo a termo
  out = out + g #acúmulo dos termos com ganho (capacitor)

plt.plot(t, x, label='Vin(t)')        #tensão de entrada
plt.plot(t, out, label='Vc(t)')       #tensão do capacitor

#detalhes do gráfico
plt.title('Tensões no circuito', fontsize=20) 
plt.xlabel('Tempo (s)', fontsize=14)
plt.ylabel('Tensão (V))',fontsize=14)
plt.xlim(right=10, left=-10)
plt.ylim(top=0.75, bottom=-.75)
plt.savefig('RC='+str(R*C)+'.png')
plt.legend()
plt.show()
```
