## (c)

$z(x,y) = x^2+y^2$

$C = x^2 + y^2$

$r = \sqrt{x^2 + y^2}$

$C = r^2$

$r = \sqrt{C}$

```python
# Imagem (contour)
# Arranjo em superfície
x, y = np.meshgrid( np.linspace(-1.5, 1.5, 100),
                    np.linspace(-1.5, 1.5, 100))
z = y**2 + x**2

plt.figure(figsize=(8,6))
plt.contourf(x,y,z, levels=30, cmap=plt.get_cmap("gray"))

C=1.5
r = np.sqrt(C)
theta = np.linspace(0, 2*np.pi, 200)
x_ = r*np.cos(theta)
y_ = r*np.sin(theta)
plt.plot(x_,y_, c='r')

plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.show()
```

# Exercício 2

$E= kq/d^2$

```python
def campo_eletrico(x0, y0, x, y, K):
  d = np.sqrt( (x-x0)**2 + (y-y0)**2 )+2
  m = K/d**2

  v = np.array([ x-x0, y-y0])
  vu = v/d

  vce = m * vu
  return vce[0], vce[1]

# Campo vetorial
x, y = np.meshgrid( np.linspace(-5, 5, 15),
                    np.linspace(-5, 5, 15))

vx0,vy0 = campo_eletrico(2, -2, x, y, 100)
vx1,vy1 = campo_eletrico(-2, 2, x, y, -100)

vx=vx0+vx1
vy=vy0+vy1

plt.figure(figsize=(8,6))
plt.quiver(x,y,vx,vy)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
