```
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```
r = 1
t = np.linspace(0, 1, 100)
x = r * np.cos(2 * np.pi * t)
y = r * np.sin(2 * np.pi * t)


t_ = np.linspace(0, 1, 20)
x_ = r * np.cos(2 * np.pi * t_)
y_ = r * np.sin(2 * np.pi * t_)
dxdt = r * 2 * np.pi * -1 * np.sin(2 * np.pi * t_)
dydt = r * 2 * np.pi * np.cos(2 * np.pi * t_)

plt.figure(figsize=(8,8))
plt.plot(x, y)
plt.quiver(x_, y_, dxdt, dydt)
plt.show()
```
